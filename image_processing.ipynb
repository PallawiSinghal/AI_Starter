```
items = os.listdir('/content')
# print (items)

for each_image in items:
  if each_image.endswith(".jpg"):
#     print (each_image)
    full_path = "/content/" + each_image
#     print (full_path)
    image = cv2.imread(full_path)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(image)
plt.colorbar()
plt.grid(False)
print (full_path)

```

```
def gamma_correction(image_data,gamma):
  print ("I am in fuction gamma correction")
  invGamma = 1.0 / gamma
  table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
  return cv2.LUT(image_data, table)
```

```
gama_input_image = image.copy()
gamma = 0.4
gama_corrected_image = gamma_correction(gama_input_image,gamma)
plt.figure()
plt.imshow(gama_corrected_image)
plt.colorbar()
plt.grid(False)
```

```
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(gama_corrected_image,'#Simmy',(187,103), font, 1,(0,255,255),2,cv2.LINE_AA)
cv2.putText(gama_corrected_image,'#Pluto',(744,534), font, 1,(0,255,255),2,cv2.LINE_AA)
cv2.putText(gama_corrected_image,'Sunday mornining 7:00 am',(86,557), font, 1,(0,255,255),2,cv2.LINE_AA)
figure_new = plt.figure()
plt.imshow(gama_corrected_image)
plt.colorbar()
plt.grid(False)
figure_new.savefig("test.png")
height_image,width_image,channel = gama_corrected_image.shape
print (height_image,width_image,channel)
```
